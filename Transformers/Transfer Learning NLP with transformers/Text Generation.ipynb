```python
from transformers import pipeline

import textwrap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pprint import pprint
```

```python
text_gen = pipeline("text-generation")
```

```python
sentence = "My name is Syed Qasim Mustafa and I like Deep Learning"  # you can put your large text as well.
text_gen(sentence)
```

```python
print(text_gen(sentence, max_length=200))
```

```python
pprint(text_gen(sentence, max_length=200)) # pretty printer lol
```

```python
pprint(text_gen(sentence, num_return_sequences=3, max_length=40))
# Text generation is random because it is generated from probability distribution
```

```python
sentence  = "I wanted to kill superman, but turns out his mother is also named Martha"
pprint(text_gen(sentence, num_return_sequences=5, max_length=40))
```
